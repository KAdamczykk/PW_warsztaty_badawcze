```python
import lmstudio as lms

input_file = 'C:/studia/3rok/WB2/homework1/homework_1_ang.txt'
output_file = 'C:/studia/3rok/WB2/homework1/odszyfrowane1.txt'

with open(input_file, 'r', encoding='utf-8') as f:
    encrypted_sentences = f.readlines()

model = lms.llm("mistral-7b-instruct-v0.3")

decrypted_sentences = []

for sentence in encrypted_sentences:
    response = model.respond(f"You are a professional caesar cipher decryptor. Solve this cipher: {sentence}")
    decrypted_text = str(response)
    decrypted_sentences.append(decrypted_text)
    print(f"Odpowiedź: {decrypted_text}")

with open(output_file, 'w', encoding='utf-8') as f:
    for decrypted in decrypted_sentences:
        f.write(decrypted + '\n')

print(f"Odszyfrowane zdania zapisano do: {output_file}")
```

```python
from together import Together

api_key_file = "C:/studia/3rok/WB2/homework1/API_KEY.txt"
with open(api_key_file, "r") as f:
    API_KEY = f.read().strip()

client = Together(api_key=API_KEY)
print("Klucz API załadowany pomyślnie!")
```

```python
def decrypt_caesar_cipher(client, sentence, idx):

    response = client.chat.completions.create(
        model="deepseek-ai/DeepSeek-R1-Distill-Llama-70B-free",
        messages=[{"role": "user", "content": f"You are a professional caesar cipher decryptor. Solve this cipher and provide the final decrypted sentence: {sentence}"}],
        max_tokens=8000,
        temperature=0.7,
        top_p=0.7,
        top_k=50,
        repetition_penalty=1.0
    )

    decrypted_text = response.choices[0].message.content.strip()

    print(f"**Zaszyfrowane ({idx}):** {sentence.strip()}")
    print(f"**Odszyfrowane ({idx}):** {decrypted_text}")
    print("-" * 80)

    return decrypted_text
```
