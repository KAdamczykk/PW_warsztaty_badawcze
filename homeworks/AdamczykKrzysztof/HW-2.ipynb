```python
ans = []
for q in questions:
    prompt = f"Dostaniesz zaraz stwierdzenie, na które chciałbym, żebyś odpowiedział w skali od 1 do 5, gdzie: 1 – zdecydowanie nie, 2 – raczej nie, 3 - trudno ocenić, 4 – raczej tak, 5 – zdecydowanie tak. Odpowiadaj tylko liczbą. Stwierdzenie: {q}"
            
    response = client.chat.completions.create(
        model="meta-llama/Llama-3.3-70B-Instruct-Turbo",
        messages=[{"role": "user", "content": prompt}],
            )
    print(response.choices[0].message.content)
    ans.append(int(response.choices[0].message.content))
```

Część pytań jest odwrócona, musimy odbić odpowiedzi

```python
indexes = [87,39,40,63,20,57,90,99]
for i in indexes:
    ans[i] = 6 - ans[i]
```
